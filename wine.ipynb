```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
base_wine = pd.read_csv('winequality-red.csv')
base_wine
```

DESCRIÇÃO DO DATASET (NENHUM DADO ESPÚRIO)

```python
base_wine.describe()
```

```python
X_wine_data = base_wine.iloc[:, 0:11].values
X_wine_data
```

```python
y_wine_data = base_wine.iloc[:, 11].values
y_wine_data
```

Normalização / Padronização dos Dados

```python
from sklearn.preprocessing import StandardScaler
scaler_wine = StandardScaler()
X_wine_data = scaler_wine.fit_transform(X_wine_data)
```

```python
X_wine_data
```

Fazendo as divisões das bases (treinamento e teste) 

```python
from sklearn.model_selection import train_test_split
```

```python
X_wine_treinamento, X_wine_teste, y_wine_treinamento, y_wine_teste = train_test_split(X_wine_data, y_wine_data ,test_size=0.20,random_state=0)
```

```python
X_wine_treinamento.shape, X_wine_teste.shape
```

```python
y_wine_treinamento.shape, y_wine_teste.shape
```

MODELO NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_wine_data = GaussianNB()
naive_wine_data.fit(X_wine_treinamento, y_wine_treinamento)
```

```python
previsoes_naive_bayes = naive_wine_data.predict(X_wine_teste)
previsoes_naive_bayes
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred=previsoes_naive_bayes, y_true=y_wine_teste)
```

MODELO ARVORE DE DECISÃO

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_wine = DecisionTreeClassifier(criterion='entropy')
tree_wine.fit(X_wine_treinamento, y_wine_treinamento)
```

```python
previsoes_arvore_unica = tree_wine.predict(X_wine_teste)
previsoes_arvore_unica
```

```python
accuracy_score(y_pred=previsoes_arvore_unica, y_true=y_wine_teste)
```

MODELO ÁRVORE DE DECISÃO (RANDOM FOREST)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest_wine = RandomForestClassifier(n_estimators=300, criterion='entropy', random_state=0)
random_forest_wine.fit(X_wine_treinamento, y_wine_treinamento)
```

```python
previsoes_random_forest_wine = random_forest_wine.predict(X_wine_teste)
previsoes_random_forest_wine
```

```python
accuracy_score(y_pred=previsoes_random_forest_wine, y_true=y_wine_teste)
```

MODELO KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_wine = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_wine.fit(X_wine_treinamento, y_wine_treinamento)
```

```python
previsoes_knn = knn_wine.predict(X_wine_teste)
previsoes_knn 
```

```python
accuracy_score(y_pred=previsoes_knn, y_true=y_wine_teste)
```

MODELO REGRESSÃO LOGÍSTICA

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_wine = LogisticRegression(random_state=1)
logistic_wine.fit(X_wine_treinamento, y_wine_treinamento)
```

```python
previsoes_logistic = logistic_wine.predict(X=X_wine_teste)
previsoes_logistic
```

```python
accuracy_score(y_pred=previsoes_logistic, y_true=y_wine_teste)
```
